```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
interval = 30
IP_change0 = np.load("/Users/eciel/Desktop/y3s2_sum/orf/summer/privacy_methods/Logistic/IF_lr/IP_change_log.npy")
IF_change = np.load("/Users/eciel/Desktop/y3s2_sum/orf/summer/privacy_methods/Logistic/IF_lr/IF_change_log.npy")
RIF_change = np.load("/Users/eciel/Desktop/y3s2_sum/orf/summer/privacy_methods/Logistic/IF_lr/RIF_change.npy")
GD_change = np.load("/Users/eciel/Desktop/y3s2_sum/orf/summer/privacy_methods/Logistic/IF_lr/GD_change.npy")
GC_change = np.load("/Users/eciel/Desktop/y3s2_sum/orf/summer/privacy_methods/Logistic/IF_lr/GC_change.npy")
K = list(range(1, IP_change0.shape[0] + 1))
K = [interval * int(i) for i in K]
```

```python
IP_change.shape
```

```python
IP_mean0 = np.mean(IP_change0, axis=1)
IF_mean = np.mean(IF_change, axis=1)
RIF_mean = np.mean(RIF_change, axis=1)
GD_mean = np.mean(GD_change, axis=1)
GC_mean = np.mean(GC_change, axis=1)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(K, IP_mean, label = "IP", linewidth=2)
plt.plot(K, IF_mean, label = "IF", linewidth=2)
plt.plot(K, RIF_mean, label = "RIF", linewidth=2)
plt.plot(K, GD_mean, label = "GD", linewidth=2)
plt.plot(K, GC_mean, label = "GC", linewidth=2)
plt.legend()
plt.xlabel("k")
plt.ylabel("mean of change of prob")
plt.title("(a) Remove trainning Points")
plt.savefig("Same.jpg")
plt.show()
```
